```python
from tkinter import *
import tkinter

top = Tk()
top.geometry('200x200')
fd = open("stockdata.txt","r")
txt = fd.read()
fd.close()
products = txt.split("\n")
Lb1 = Listbox(top)
for product in products:
    prod = product.split(",")
    Lb1.insert(1,prod)


Lb1.pack()
top.mainloop()


fd = open("stockdata.txt","r")
txt = fd.read()
fd.close()

products = txt.split("\n")

ui_prod  = str(input("Enter the product_Id: "))
ui_quant = input("Enter the quantity: ")

for product in products:
    prod = product.split(",")
    
    
    if(prod[0] == ui_prod):
        print("***********************")
        print("Product ID: ", prod[0])
        print("Name: ", prod[1])
        print("Price: ",prod[2])
        print("Quant: ",ui_quant)
        print("------------------------")
        print("Billing Amout: ", int(ui_quant) * int(prod[2]))
        print("***********************")

        
new_record = []

for product in products:
    prod = product.split(",")
    if(ui_prod == prod[0]):
        prod[3] = str(int(prod[3]) - int(ui_quant))
        
    new_record.append(prod[0] + "," + prod[1] + "," + prod[2] + "," + prod[3] + "\n")
    
new_record[-1] = new_record[-1][:-1]



fd = open("stockdata.txt", 'w')

for i in new_record:
    fd.write(i)
    
fd.close()


new_record2 = []
fd = open("selldata.txt","r")
txt = fd.read()
fd.close()

products2 = txt.split("\n")

for product2 in products2:
    prod = product2.split(",")   
    if(ui_prod == prod[0]):
        prod[3] = str(int(prod[3]) + int(ui_quant))       
    new_record2.append(prod[0] + "," + prod[1] + "," + prod[2] + "," + prod[3] + "\n")
    
new_record2[-1] = new_record2[-1][:-1]



fd = open("selldata.txt", 'w')

for i in new_record2:
    fd.write(i)
    
fd.close()
```
